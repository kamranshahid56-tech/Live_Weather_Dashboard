```python
import requests
import pandas as pd
import plotly.express as px

#configuration
API_KEY = "4988e929a056e2bc288e7e33a361f374"
CITY = input("Enter City: ")
URL = f"https://api.openweathermap.org/data/2.5/forecast?q={CITY}&appid={API_KEY}&units=metric"
```

```python
# Fetch Data
response = requests.get(URL)
data = response.json()

# Check if the request was successful
if response.status_code == 200:
    forecast_list = []
    for entry in data['list']:
        forecast_list.append({
            "Time": entry['dt_txt'],
            "Temp": entry['main']['temp'],
            "Condition": entry['weather'][0]['main']
        })
    df = pd.DataFrame(forecast_list)
    print("Data successfully retrieved!")
else:
    print(f"Error {response.status_code}: {data.get('message', 'Unknown error')}")
```

```python
df = pd.DataFrame(forecast_list)
fig = px.line(df,
              x='Time',
              y='Temp',
              color='Condition', # This adds the condition to the legend and colors the data
              title=f"5-Day Temperature Forecast for {CITY}",
              labels={'Temp': 'Temperature (째C)', 'Time': 'Date/Time'},
              markers=True,
              hover_data={'Condition': True} # Ensures condition shows when you hover
             )
fig.show()
```
